```python
def mat_mul(x, y):
    n = len(x)
    k = len(y)
    m = len(y[0])
    z = [[0 for _ in range(m)] for _ in range(n)]
    for i in range(n):
        for j in range(m):
            for l in range(k):
                z[i][j] += x[i][l] * y[l][j]
    return z

def mat_mul_n(mat, n):
    t = len(mat)
    A = [[0 for _ in range(t)] for _ in range(t)]
    for i in range(t):
        A[i][i] = 1
    while n > 0:
        if n % 2 == 1:
            A = mat_mul(A, mat)
        mat = mat_mul(mat, mat)
        n //= 2
    return A

print("---- win rate ----")

def solve_win(x, y, p):
    n = x + y
    
    # x_0 = 0, x_n = 1
    # x_i = p * x_{i+1} + (1-p) * x_{i-1}
    # x_{i+1} = (p-1) / p * x_{i-1} + 1 / p * x_i
    mat = [[0, 1], [(p-1.0)/p, 1/p]]
    A = mat_mul_n(mat, n)
    B = mat_mul_n(mat, x)
    print(A,B)
    
    # A * [x_0, x_1]^T = [x_n, x_{n+1}]^T
    # A * [0, x_1]^T = [1, x_{n+1}]^T
    # A[0][0] * 0 + A[0][1] * x_1 = 1
    # x_1 = 1 / A[0][1]
    # ans = B[0][1] * x_1
    # ans = B[0][1] / A[0][1]
    ans = B[0][1] / A[0][1]
    
    win = [0 for _ in range(n+1)]
    win[0] = 0.0
    win[1] = 1 / A[0][1]
    for i in range(2, n+1):
        win[i] = win[i-1] / p - (1-p)/p * win[i-2]
    #print(win)
    
    return ans

print(solve_win(50,50,0.5))
print(solve_win(1,99,0.7))

print("---- step ----")

def solve_step(x, y, p):
    n = x + y
    
    # x_0 = 0, x_n = 0
    # x_i = p * x_{i+1} + (1-p) * x_{i-1} + 1
    # x_{i+1} = (p-1) / p * x_{i-1} + 1 / p * x_i - 1 / p
    mat = [ [0, 1, 0], [(p-1.0)/p, 1/p, -1/p], [0, 0, 1]]
    A = mat_mul_n(mat, n)
    B = mat_mul_n(mat, x)
    print(A,B)
    
    # A * [x_0, x_1, 1]^T = [x_n, x_{n+1}, 1]^T
    # A * [0, x_1, 1]^T = [0, x_{n+1}, 1]^T
    # A[0][0] * 0 + A[0][1] * x_1 + A[0][2] = 0
    # x_1 = -A[0][2] / A[0][1]
    # ans = B[0][1] * x_1 + B[0][2]
    # ans = B[0][2] - A[0][2] * B[0][1] / A[0][1]
    ans = B[0][2] - A[0][2] * B[0][1] / A[0][1]
    step = [0 for _ in range(n+1)]
    step[0] = 0
    step[1] = -A[0][2] / A[0][1]
    for i in range(2, n+1):
        step[i] = step[i-1] / p - (1-p)/p * step[i-2] - 1 / p
    print(step)
    
    return ans

print(solve_step(10,10,0.5))
print(solve_step(2,2,0.6))
```
